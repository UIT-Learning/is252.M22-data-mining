```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
%matplotlib inline
```

```python
#c vo file nh cn gim v kch thc hin th ra mn hnh
image = Image.open('dog.png')
axes = plt.axes(xticks=[], yticks=[])
axes.imshow(image)
```

```python
#Chuyn nh c c thnh mng s v ly thng tin chiu rng, cao ca nh
imarr = np.array(image)
rows = imarr.shape[0]
cols = imarr.shape[1]
```

```python
#Bin i cc gi tr mu t on [0...255] v on [0...1] v chuyn dng mng t 3 chiu (cao, rng, su) thnh 2 chiu (cao x rng, su)
data = imarr / 255.0
data = data.reshape(rows * cols, 3)
data.shape
```

```python
#Chy thut ton gom cm vi s mu mong mun l 32. Tip theo , ly ra mng biu din hnh sau khi  gom cm
kmeans = KMeans(n_clusters=32, max_iter=300)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]
```

```python
#Khi phc dng mng va thu c t 2 chiu thnh 3 chiu nh ban u
img_recolored = new_colors.reshape(imarr.shape)
```

```python
#Biu din hnh kt qu v so snh vi nh gc ban u
fig, ax = plt.subplots(1, 2, figsize=(16, 6), subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace = 0.05)
ax[0].imshow(image)
ax[0].set_title('Original Image', size = 16)
ax[1].imshow(img_recolored)
ax[1].set_title('32-color Image', size = 16)
```

```python
#Lu hnh kt qu xung  a cng
img_recolored = img_recolored * 255
new_img = Image.fromarray(img_recolored.astype(np.uint8))
new_img.save('compressed_img2.png')
```

```python
#Xem kt qu
image = Image.open('compressed_img2.png')
axes = plt.axes(xticks=[],yticks=[])
axes.imshow(image)
```

```python
image = Image.open('tiger.png')
axes = plt.axes(xticks=[],yticks=[])
axes.imshow(image)
```

```python
#Chuyn hnh nh thnh mng  ly chiu rng v chiu cao
imarr =np.array(image)
rows = imarr.shape[0]
cols = imarr.shape[1]
```

```python
#Bin i cc gi tr mu t on [0...255] thnh on [0..1] v chuyn dng mng t 3 -->2
data = imarr
data = data.reshape(rows*cols,3)
data.shape
```

```python
#chy thut ton gom cm
kmeans = KMeans(n_clusters = 32, max_iter =300)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]
```

```python
#Khi phc dng mng va thu c t 2 chiu thnh 3 chiu nh ban u
img_recolored = new_colors.reshape(imarr.shape)
```

```python
#Biu din so snh 2 hnh

fig,ax = plt.subplots(1,2,figsize=(16,6),subplot_kw = dict(xticks =[],yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(image)
ax[0].set_title("Hnh gc",size=16)

ax[1].imshow(np.uint8(img_recolored))
ax[1].set_title("Hnh 32-color",size=16)
```

```python
#Lu kt qu
img_recolored = img_recolored*255
new_img = Image.fromarray(img_recolored.astype(np.uint8))
new_img.save('tiger_new2.png')
```

```python
image = Image.open('tiger_new2.png')
axes = plt.axes(xticks=[],yticks=[])
axes.imshow(image)
```
