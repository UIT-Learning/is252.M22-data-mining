```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
data = pd.read_csv('C:\\Users\\84346\\Desktop\\Data\\winequality-red.csv')
data
```

```python
data.info()
```

```python
X = data.drop('quality', axis=1)
y = data.quality
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
features = data.drop('quality', axis=1)
labels = data['quality']
```

```python
features
```

```python
labels
```

```python
features.select_dtypes(exclude=["int64"]).columns
```

```python
features_onehot =  pd.get_dummies(features, columns=features.select_dtypes(exclude=["int64"]).columns)
features_onehot
```

```python
X_train = features_onehot.iloc[:len(X_train)]
X_test = features_onehot.iloc[len(X_train):]
y_train = labels.iloc[:len(X_train)]
y_test = labels.iloc[len(X_train):]
```

#### criterion="entropy"

```python
clf = tree.DecisionTreeClassifier(criterion="entropy", random_state=0)
clf.fit(X_train, y_train)
```

```python
tree_pred = clf.predict(X_test)

tree_score = metrics.accuracy_score(y_test, tree_pred)

print("Decision Tree Accuracy: ", tree_score)
print("Decision Tree Classification Report: ", metrics.classification_report(y_test, tree_pred))
```

```python
tree_cm = metrics.confusion_matrix(y_test, tree_pred)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(tree_cm, annot=True, fmt=".3f", linewidths=.5, square=True, cmap='Blues_r')
plt.ylabel('Actual')
plt.xlabel('Predicted')
title = "Decision Tree Accuracy score {0}".format(tree_score)
plt.title(title, size=15)
```

```python
fig, ax = plt.subplots(figsize=(50, 24))
tree.plot_tree(clf, filled=True, fontsize=10)
plt.savefig("decision_tree", dpi=100)
plt.show()
```

#### criterion=’gini’.

```python
clf1 = tree.DecisionTreeClassifier(criterion="gini", random_state=0)
clf1.fit(X_train, y_train)
```

```python
tree_pred = clf1.predict(X_test)

tree_score = metrics.accuracy_score(y_test, tree_pred)

print("Decision Tree Accuracy: ", tree_score)
print("Decision Tree Classification Report: ", metrics.classification_report(y_test, tree_pred))
```

```python
tree_cm = metrics.confusion_matrix(y_test, tree_pred)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(tree_cm, annot=True, fmt=".3f", linewidths=.5, square=True, cmap='Blues_r')
plt.ylabel('Actual')
plt.xlabel('Predicted')
title = "Decision Tree Accuracy score {0}".format(tree_score)
plt.title(title, size=15)
```

```python
fig, ax = plt.subplots(figsize=(50, 24))
tree.plot_tree(clf1, filled=True, fontsize=10)
plt.savefig("decision_tree", dpi=100)
plt.show()
```

### Naive Bayes

```python
gnb = GaussianNB()
bayes_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
bayes_score = metrics.accuracy_score(y_test, bayes_pred)
print("Naive Bayes Accuracy: ", bayes_score)
print("Naive Bayes Classification Report: ", metrics.classification_report(y_test, bayes_pred))
```

```python
bayes_cm = metrics.confusion_matrix(y_test, bayes_pred)
plt.figure(figsize=(12, 12))
sns.heatmap(bayes_cm, annot=True, fmt='.3f', linecolor='white', linewidths=.5, square=True, cmap='Greens')
plt.ylabel('Actual')
plt.xlabel('Predicted')
title = "Naive Bayes Accuracy score {0}".format(bayes_score)
plt.title(title, size=15)
```
